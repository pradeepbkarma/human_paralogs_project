```python
# load classes and functions and packages 
import pandas as pd 
from get_novel_network_clusters import filter_df, create_louvain_graph, plot_graph
```

```python
''' 
Network clusters on paralogs found by integrated homolog searches and 
structural comparisons using TM-scores. Here, we use kinases 

'''

# load the data 
kinase_data = pd.read_csv("../data/kinase_paralogs.tsv", sep="\t")
# ser_thr_data = pd.read_csv("../data/ser_thr_df.tsv", sep="\t")
kinase_ref = list(kinase_data['reference'].unique())

filtered_df, novel_list = filter_df(kinase_data, kinase_ref)

G, large_partition, small_partition = create_louvain_graph(filtered_df)
graph = plot_graph(G, large_partition, novel_list, "../data/", 'kinase_novel_clusters')
```
